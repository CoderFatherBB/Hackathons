```python
from ultralytics import YOLO

# Load a YOLOv8 classification model (choose 'n', 's', 'm', 'l', or 'x' for the model size)
model = YOLO("yolov8s-cls.pt")  # YOLOv8 Nano for classification

# Train the model
model.train(
    data="/content/Final",  # Path to your processed dataset folder
    epochs=20,                # Number of epochs
    # patience=10,               # Patience for early
    imgsz=256,                 # Input image size
    batch=16,                  # Batch size
    # device="mps",              # Use Apple Metal for faster training
    project="Tomato Disease",   # Save results in the "BlurDetection" folder
    name="tomo",        # Name of the experiment
    amp=True,                  # Automatic mixed precision - faster and uses less memory
    workers=4                  # Number of data loader workers
)
```
