```python
import os
import shutil
from sklearn.model_selection import train_test_split

def split_dataset(source_dir, target_dir, split_ratio=0.8):
    for class_name in ['Bud Root Dropping', 'Bud Rot','Gray Leaf Spot','Leaf Rot','Stem Bleeding']:
        class_dir = os.path.join(source_dir, class_name)
        images = os.listdir(class_dir)

        train_images, val_images = train_test_split(images, train_size=split_ratio, random_state=42)

        for split, split_images in [('train', train_images), ('val', val_images)]:
            split_dir = os.path.join(target_dir, split, class_name)
            os.makedirs(split_dir, exist_ok=True)
            for img in split_images:
                shutil.copy(os.path.join(class_dir, img), os.path.join(split_dir, img))

# Define source and target directories
source_dir = "/content/Coconut Tree Disease Dataset"
target_dir = "/content/Final"

# Split the dataset
split_dataset(source_dir, target_dir)
print("Dataset split into train and val folders!")
```
